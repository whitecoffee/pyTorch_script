```python
import pandas as pd
import numpy as np

landmarks_frame = pd.read_csv('Try.csv')
landmarks = landmarks_frame.as_matrix().astype('float')
```

```python
a = landmarks.reshape(-1, 90, 3)
#print(a[1])
```

```python
import scipy.io

matfile = 'plot3.mat'

scipy.io.savemat(matfile, mdict={'out': a}, oned_as='row')

matdata = scipy.io.loadmat(matfile)

assert np.all(a == matdata['out'])
```

```python
from mpl_toolkits.mplot3d import Axes3D  
import matplotlib.pyplot as plt 
#生成画布 
figure=plt.figure() 
ax=figure.gca(projection='3d') 
#生成向量 
x = a[1, :, 0]
y = a[1, :, 1]
z = a[1, : ,2]

x1 = a[2, :, 0]
y1 = a[2, :, 1]
z1 = a[2, : ,2]
fig1 = ax.plot(x,y,z) 
ax.plt
plt.show()
```
